# Normal Distribution

- Normal distribution, also known as Gaussian distribution, is one of the most important probability distributions in statistics. It has several key properties:

 1. Bell Shape: It exhibits a symmetric, bell-shaped curve when plotted on a graph. This shape is characterized by a single peak at the mean, with the values decreasing symmetrically as you move away from the mean in both directions.

 2. Symmetry: Normal distributions are perfectly symmetric about their mean. This means that the probability of observing a value to the left of the mean is equal to the probability of observing the same value to the right of the mean.

 3. Mean: The mean (μ) of a normal distribution represents its central tendency. It is the arithmetic average of all the data points and serves as the midpoint of the distribution. The peak of the bell-shaped curve occurs at the mean.

 4. Standard Deviation: The standard deviation (σ) measures the dispersion or spread of the data points around the mean. It quantifies the average distance between each data point and the mean. A smaller standard deviation indicates that the data points are closer to the mean, while a larger standard deviation indicates greater variability.

 5. Probability Density Function: The probability density function (PDF) of a normal distribution is described by the familiar bell-shaped curve formula, which determines the likelihood of observing a particular value. It is defined by the mean and standard deviation of the distribution.

##	Plotting Normal Distribution

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random

# Generate random data following a normal distribution using Pandas
random.seed(0)  # For reproducibility
data = pd.DataFrame({'value': [random.gauss(0, 1) for _ in range(1000)]})  # Mean=0, Standard deviation=1, 1000 data points

# Plot the histogram using Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(data['value'], kde=True, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Data following Normal Distribution', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Density plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random

# Generate random data following a normal distribution using Pandas
random.seed(0)  # For reproducibility
data = pd.DataFrame({'value': [random.gauss(0, 1) for _ in range(1000)]})  # Mean=0, Standard deviation=1, 1000 data points

# Plot the density plot using Seaborn
plt.figure(figsize=(8, 6))
sns.kdeplot(data['value'], color='skyblue', linewidth=2)
plt.title('Density Plot of Random Data following Normal Distribution', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## Interpreting Normal Distribution Plots: 

- We generate random data following a normal distribution using the random module as before.
  
- We create a DataFrame named data with one column named 'value' containing 1000 data points.
  
- We plot the density plot of the generated data using Seaborn's sns.kdeplot() function, specifying the data and color. We also specify other plot parameters such as linewidth, and add labels, title, and grid lines.
  
- The plot demonstrates the smoothed density estimation of the distribution of the randomly generated data following a normal distribution.

## 	Checking Normality

```python
from scipy.stats import kstest
from statsmodels.graphics.gofplots import qqplot
import matplotlib.pyplot as plt

# Perform Kolmogorov-Smirnov test
ks_statistic, p_value = kstest(data['value'], 'norm')

# Interpret the results
alpha = 0.05
if p_value > alpha:
    print("The data follows a normal distribution (fail to reject H0)")
else:
    print("The data does not follow a normal distribution (reject H0)")

# Generate the Q-Q plot
qqplot(data['value'], line='s')  # 's' indicates a standardized line (slope = 1, intercept = mean)
plt.title('Q-Q Plot of Random Data')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```
